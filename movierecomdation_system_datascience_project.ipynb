A movie recommendation system is uses machine learning techniques, such as collaborative filtering and content-based filtering, to predict what a user will enjoy, leading to more relevant suggestions. Platforms like Netflix and YouTube use these systems to retain users by offering personalized content and enhancing their overall viewing experience.


```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('/content/tmdb_5000_movies.csv')
credits=pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.sample()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast

# Now you can use ast.literal_eval()
string_data = "[1, 2, {'key': 'value'}]"
parsed_data = ast.literal_eval(string_data)
print(parsed_data)
print(type(parsed_data))
```

```python
type(string_data)
```

```python
type(movies.loc[0,'genres'])
```

```python
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast']
```

```python
def convert4(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(convert4)
```

```python
movies.sample(5)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head(5)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head(5)
```

```python
newdf=movies[['movie_id','title','tags']]
newdf.head(5)
```

```python
newdf['tags']=newdf['tags'].apply(lambda x:" ".join(x))
```

```python
newdf['tags']=newdf['tags'].apply(lambda x:x.lower())
```

```python
newdf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=count.fit_transform(newdf['tags']).toarray()
```

```python
vectors[0]
```

```python
count.get_feature_names_out()
```

```python
len(count.get_feature_names_out())
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
newdf['tags']=newdf['tags'].apply(stem)
```

```python
count.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim=cosine_similarity(vectors)
```

```python
cos_sim[1]
```

```python
newdf.head()
```

```python
def recommended_movie(similar_movie):

    index=newdf[newdf['title']==similar_movie].index[0]
    similar_movies=sorted(list(enumerate(cos_sim[index])),reverse=True,key=lambda x:x[1])[1:6]
    print(f'Recommaned movie of {similar_movie} are:')
    print('*'*27)
    for movie in similar_movies:
      print(newdf.iloc[movie[0]].title)

```

```python
recommended_movie('Autumn in New York')
```
